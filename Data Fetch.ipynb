1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x = tips["total_bill"],y = tips["tip"], hue = tips["sex"], style = tips["smoker"], size = tips["size"]) #Scatter plot with multiple variables  #hue, style, size  #hue = color, style = marker, size = size of marker 
```

2. Bar plot (Numwrical vs Categorical)

```python
sns.barplot(x = titanic["pclass"], y = titanic["age"])
```

```python
sns.barplot(x = titanic["pclass"], y = titanic["fare"], hue = titanic["sex"])
```

3. Box plot (Numerical vs Categorical)

```python
sns.boxplot(x = titanic["sex"], y = titanic["age"], hue = titanic["pclass"])
```

4. Distplot (Numerical vs categorical)

```python
sns.displot(titanic["age"], kde = True) #Histogram
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(titanic[titanic["survived"] == 1]["age"], kde=True)
sns.displot(titanic[titanic["survived"] == 0]["age"], kde=True)
```
