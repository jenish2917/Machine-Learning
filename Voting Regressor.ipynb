```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('boston_house_prices.csv')
```

```python
df
```

```python
df.columns = df.iloc[0]
```

```python
df
```

```python
df = df.drop(0)
```

```python
df
```

```python
x = df.drop('MEDV', axis=1)
y = df['MEDV']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [
    ('lr', lr),
    ('dt', dt),
    ('svr', svr)
]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], x, y, cv=10, scoring='r2')
    print(f'{estimator[0]}: {scores.mean()}')
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr, x, y, cv=10, scoring='r2')
print(f'Voting Regressor: {scores.mean()}')
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        for k in range(1, 4):
            weights = [i, j, k]
            vr = VotingRegressor(estimators=estimators, weights=weights)
            scores = cross_val_score(vr, x, y, cv=10, scoring='r2')
            print('For i = {}, j = {}, k = {}: {}'.format(i, j, k, np.round(np.mean(scores), 2)))
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
df2 = DecisionTreeRegressor(max_depth=3)
df3 = DecisionTreeRegressor(max_depth=5)
df4 = DecisionTreeRegressor(max_depth=7)
df5 = DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [
    ('dt1', dt1),
    ('df2', df2),
    ('df3', df3),
    ('df4', df4),
    ('df5', df5)
]
```

```python
for estimetor in estimators:
    scores = cross_val_score(estimetor[1], x, y, cv=10, scoring='r2')
    print(f'{estimetor[0]}: {scores.mean()}')
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr, x, y, cv=10, scoring='r2')
print(f'Voting Regressor: {scores.mean()}')
```
